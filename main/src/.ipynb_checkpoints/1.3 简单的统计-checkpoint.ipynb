NLTK中频率分布的函数

|Example|Description|
|--|--|
|fdist = FreqDist(samples)|使用提供的样本产生它的频率分布|
|fdist = inc(sample)|给这个样本再增加一部分样本|
|fdist[monstrous]|查询指定的单词的出现次数|
|fdist.N()|样本总共的数量|
|fdist.freq("monstrous")|计算一个给定样本的频率|
|fdist.keys()|根据频率降低顺序对样本进行排序|
|fdist.max()|计算样本中出现的最大的次数|
|fdist.plot()|画出样本的频率分布|
